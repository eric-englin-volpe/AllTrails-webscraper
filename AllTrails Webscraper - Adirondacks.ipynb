```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
def get_all_hikes(browser, site):
    browser.get(site)
    while True:
        try:
            time.sleep(3)
            browser.find_element_by_class_name("styles-module__button___1nuva").click()
        except:
            break
    soup = BeautifulSoup(browser.page_source, 'lxml')
    return soup

```

```python
def parse_data(soup):
    #find regional website
    #find regional name
    region_title_list = []
    region_site_list =  []
    for x in soup.body.findAll("a"):
        if "xlate-none styles-module__location___11FHK styles-module__info___1Mbn6 styles-module__link___3T9FO" in str(x):
            #print(x.contents[0])
            try:
                region_site_list.append("https://www.alltrails.com" + x.get('href'))
            except:
                region_site_list.append("No Site")
            region_title_list.append(x.get('title'))
    
    park_title_list = []
    park_site_list = []
    for x in soup.body.findAll("a"):
        if "xlate-none styles-module__link___12BPT" in str(x):
            #print(x.contents[0])
            try:
                park_site_list.append("https://www.alltrails.com" + x.get('href'))
            except:
                park_site_list.append("No Site")
            park_title_list.append(x.get('title'))

    num_ratings_list = []
    for x in soup.body.findAll("span"):
        if "styles-module__count___2pQCU" in str(x):
            rating = x.contents[0]
            rating = rating[1:-1]
            num_ratings_list.append(rating)

    num_stars_list = []
    for x in soup.body.findAll("span"):
        if "MuiRating-root default-module__rating___1k45X MuiRating-sizeLarge MuiRating-readOnly" in str(x):
            num_stars_list.append(x.get("aria-label"))

    for y in num_stars_list:
        if "0 Stars" in str(y):
            num_ratings_list.append(0)
            
    length_list = []
    for x in soup.body.findAll("span"):
        if "Length" in str(x):
            if len(x.contents)>1:
                length_list.append(x.contents[-1])
            else:
                length_list.append(x.contents[0][x.contents[0].find(":")+2:])

    difficulty_list = []
    for x in soup.body.findAll("span"):
        if "module__selected___3fawg" in str(x):
            if len(x.contents)>1:
                difficulty_list.append(x.contents[0])
            else:
                difficulty_list.append("None found")

                
    est_time_list = []
    for x in soup.body.findAll("span"):
        if "12.9 mi" in str(x) and region_title_list[0]=="High Peaks Wilderness": #one park doesn't have an estimated time
            est_time_list.append(0)
        if "Est." in str(x):
            if len(x.contents)>1:
                est_time_list.append(x.contents[-1])
            else:
                est_time_list.append(x.contents[0])
                
    print(len(park_title_list), len(park_site_list), len(region_title_list), len(region_site_list), \
            len(num_ratings_list), len(num_stars_list), len(length_list), len(est_time_list), len(difficulty_list))
    return park_title_list, park_site_list, region_title_list, region_site_list, num_ratings_list, num_stars_list, length_list, est_time_list, difficulty_list
```
